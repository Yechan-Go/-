* Item

####

* ex)

```python
#  
a = [5, 10, 15, 20, 25, 30, 35, 40]

# 
print("a[2] =", a[2])  # Output: a[2] = 15

# 
print("a[0:3] =", a[0:3])  # Output: a[0:3] = [5, 10, 15]
print("a[5:] =", a[5:])  # Output: a[5:] = [30, 35, 40]

#  
a[0] = 100
print(a)  # Output: [100, 10, 15, 20, 25, 30, 35, 40]
```

### (set)
*       
*  {}       
*      Slicing operation     
*    (list )     
* add(), update(), remove(), discard()       

```python
# (set)
my_set = {1, 2}
print(my_set)  # : {1, 2}

#  
my_set.add(3)
print(my_set)  # : {1, 2, 3}

#   
my_set.update([4, 5], {6, 8})
print(my_set)  # : {1, 2, 3, 4, 5, 6, 8}

#    
a = [1, 2, 2, 3, 3, 3]
print(a)  # : [1, 2, 2, 3, 3, 3]
a = set(a)
print(a)  # : {1, 2, 3}
```
