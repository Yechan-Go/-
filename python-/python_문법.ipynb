##01.명령문(statement)  
###할당문(Assignment Statement) 
* 변수를 할당하는 문장  
###다중 라인 명령문(Multi-line Statement)  
* 역슬래시(\)나 괄호((), [], {})를 사용해서 여러 줄에 걸쳐 명령문 작성 가능  
###다중 명령문(Multiple Statements)  
* 세미콜론(;)으로 한 줄에 여러 명령문 작성 가능

```python
#01.명령문(statement)
#할당문(Assignment Statement)
a_01= 1

#다중 라인 명령문(Multi-line Statement)
b_01 = 1 + 2 + 3 + \
  4 + 5 + 6 + 7
print(f'b_01 = {b_01}')
#b_01 = 28

#다중 명령문 (Multiple Statements)
c_01 = 10; d_01 = 20; e_01 = 30
```

  
##02.들여쓰기, 주석, 독스트링(indentation, comments, docstring)  
###들여쓰기(indentation)  
* 코드 블록을 들여쓰기로 구분  
* ex) for 루프나 if 문 안의 코드는 들여쓰기를 통해 그 블록에 속함을 확인 가능  
###주석(comments)  
* 한 줄 주석은 # 기호, 여러 줄 주석은 '...' or "..."사용  
###독스트링(docstring)  
* 함수, 메서드, 클래스, 모듈 정의 후에 사용하는 문자열 리터럴  
* __doc__ 속성을 통해 객체와 연결

```python
#02.들여쓰기, 주석, 독스트링(indentation, comments, docstring)
#들여쓰기(indentation)
for i in range(1,11):
  print('i = ',i)
  if i == 5:
    break
  
print()

if True:
  print('hello')

if True: print('hello')

print()

#독스트링(docstring)  
def double(num):
  """입력값의 두배수를 반환"""
  return 2*num
print(double.__doc__)
```

  
### 리스트(list)  
* Item 의 순서가 지정된 목록  
* 대괄호 '[ ]' 안에 아이템을 쉼표로 구분하여 나열  
* 리스트 내부의 아이템은 0부터 시작하는 인덱스로 접근  
#### 슬라이싱  
* 리스트 아이템의 부분 집합을 가져올 수 있음
* 아이템 변경 가능
* ex)  
a[0] = 100은 a의 첫 번째 아이템을 100으로 변경


```python
# 리스트 선언
a = [5, 10, 15, 20, 25, 30, 35, 40]

# 인덱싱
print("a[2] =", a[2])  # Output: a[2] = 15

# 슬라이싱
print("a[0:3] =", a[0:3])  # Output: a[0:3] = [5, 10, 15]
print("a[5:] =", a[5:])  # Output: a[5:] = [30, 35, 40]

# 아이템 변경
a[0] = 100
print(a)  # Output: [100, 10, 15, 20, 25, 30, 35, 40]
```

  
### 튜플(tuple)  
* 리스트와 기능이 거의 동일하지만, 선언 이후에 변경 불가능  
* ex  
t가 튜플일때, t[0] = 10은 오류발생
* 일반적으로 리스트보다 연산 속도가 빠름  
* 괄호 () 안에 쉼표로 구분하여 아이템을 나열 

```python
# 튜플 선언
t = (5, 'program', 1+3j)

# 아이템 접근
print("t[1] =", t[1])  # 출력: t[1] = program

print("t[0:3] =", t[0:3])  # 출력: t[0:3] = (5, 'program', (1+3j))

# 아래 코드는 오류발생
# t[0] = 10 
```

  
### 문자열(string)  
* 유니코드(Unicode) 문자들의 나열(sequence)  
* 슬라이싱 연산자 사용해서 접근 가능하지만 문자열 자체를 변경할 수는 없음
* enumerate() 함수를 사용해서 문자열의 인덱스와 값(글자)의 쌍을 생성할 수 있음

```python
# 문자열 선언
s = "This is a string"
str = 'cold'

# 문자열 출력
print(s)  # 출력: This is a string

# 여러 줄 문자열 선언
s = '''A multiline
string'''

print(s)  # 출력: A multiline\nstring

# enumerate()를 사용하여 인덱스와 값의 쌍 생성
list_enumerate = list(enumerate(str))
print('list(enumerate(str)) =', list_enumerate)  # 출력: list(enumerate(str)) = [(0, 'c'), (1, 'o'), (2, 'l'), (3, 'd')]

# 문자열 길이 출력
print('len(str) =', len(str))  # 출력: len(str) = 4

# 문자열 인덱스로 접근
print("s[4] =", s[4])  # 출력: s[4] = o
print("s[6:11] =", s[6:11])  # 출력: s[6:11] = world

# 문자열은 불변이므로, 아래 코드는 오류를 발생시킵니다.
# s[5] = 'd'  # 오류 발생
```

### 집합(set)
* 고유한 아이템들의 정렬되어있지 않은 모음  
* 중괄호 {} 안에 쉼표로 구분하여 아이템을 나열  
* 중복된 아이템을 가질 수 없고 Slicing operation을 통한 접근 불가  
* 변경 가능한 아이템 (list 등)은 가질 수 없음  
* add(), update(), remove(), discard() 메서드를 사용하여 아이템을 추가하거나 제거할 수 있음

```python
# 집합(set)
my_set = {1, 2}
print(my_set)  # 출력: {1, 2}

# 아이템 추가
my_set.add(3)
print(my_set)  # 출력: {1, 2, 3}

# 여러 아이템 추가
my_set.update([4, 5], {6, 8})
print(my_set)  # 출력: {1, 2, 3, 4, 5, 6, 8}

# 아이템이 중복될 수 없음
a = [1, 2, 2, 3, 3, 3]
print(a)  # 출력: [1, 2, 2, 3, 3, 3]
a = set(a)
print(a)  # 출력: {1, 2, 3}
```
